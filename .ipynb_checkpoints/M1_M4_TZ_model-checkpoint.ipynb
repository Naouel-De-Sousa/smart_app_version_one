# filter data on TZ_AC_MLGL_max <= 27700 

```python
data_fitered_MLGL = data.loc[data["TZ_AC_MLGL_max"]<=27700]
```

```python
data_fitered_MLGL.info()
```

## check the result

```python
sns.displot(data_fitered_MLGL,x = "TZ_AC_MLGL_correl_tx", 
            hue=data_fitered_MLGL["TZ_AC_MLGL_correl_tx"]>=20000)
```

```python
X = data_fitered_MLGL[['NZ1','PITCH_1','ROLL_1','Vx_GRD','Delta_Ny_AC','Wy_AC', 'Wx_AC','time_nz_max',
        'delta_pitch', 'delta_roll','delta_nz']]
```

```python
y = data_fitered_MLGL[["TZ_AC_MLGL_correl_tx"]]
```

## split and scale the data to fit new model

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .20, random_state= 42)
```

```python
scaler = RobustScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# grid search cv

```python
grid_search = {
    "n_estimators":[100,500],## Number of trees in random forest
    "max_depth":[30,40,50] ,# maximum number of level in tree
    "max_features":["auto","sqrt"] # methode of selectiong samples of each tree
}

rf_grid_search = RandomForestRegressor()

rf_grid_search= GridSearchCV(estimator=rf_grid_search,
                            param_grid=grid_search,
                            cv=5,
                            n_jobs=-1,
                            verbose=2)
```

```python
rf_grid_search.fit(X_train,y_train)
```

```python
rf_grid_search.best_estimator_
```

```python
rf_grid_search.best_score_
```

```python
rf_grid_search.best_params_
```

### best model

```python
model_TZ_AC_MLGL_correl_tx = RandomForestRegressor(max_depth=40, n_estimators=500,max_features = 'sqrt')
```

```python
model_TZ_AC_MLGL_correl_tx.fit(X_train,y_train)
```

## metrics

```python
y_test_pred = model_TZ_AC_MLGL_correl_tx.predict(X_test)
y_train_pred = model_TZ_AC_MLGL_correl_tx.predict(X_train)
```

```python
from sklearn import metrics

print('Mean Absolute Error (MAE):', metrics.mean_absolute_error(y_test, y_test_pred))
print('Mean Squared Error (MSE):', metrics.mean_squared_error(y_test, y_test_pred))
print('Root Mean Squared Error (RMSE):', metrics.mean_squared_error(y_test, y_test_pred, squared=False))
print('Mean Absolute Percentage Error (MAPE):', metrics.mean_absolute_percentage_error(y_test, y_test_pred))
print('Explained Variance Score:', metrics.explained_variance_score(y_test, y_test_pred))
print('Max Error:', metrics.max_error(y_test, y_test_pred))
#print('Mean Squared Log Error:', metrics.mean_squared_log_error(y_test, y_test_pred))
print('Median Absolute Error:', metrics.median_absolute_error(y_test, y_test_pred))
print('R^2:test', metrics.r2_score(y_test, y_test_pred))
print('R^2:train', metrics.r2_score(y_train, y_train_pred))
```

```python
plt.figure(figsize=(16,8))
plt.scatter(y_test, y_test_pred, edgecolors=(0,0,0),label = 'y_test')
#plt.scatter(y_train, y_train_pred, edgecolors=(1,0,0),label = 'y_train')

plt.plot([y.min(),y.max()],[y.min(),y.max()],lw=4)
plt.xlabel("true value")
plt.ylabel("Predicted")
plt.title("TZ_AC_MLGL_correl_tx")
plt.legend(labels = ["true value","test"])
```

```python
plt.figure(figsize=(16,8))
plt.scatter(y_train, y_train_pred, edgecolors=(1,0,0),label = 'y_train')
plt.plot([y.min(),y.max()],[y.min(),y.max()],lw=4)
plt.xlabel("true value")
plt.ylabel("Predicted")
plt.title("TZ_AC_MLGL_correl_tx")
plt.legend(labels = ["true value","train"])
```

```python
plt.figure(figsize=(7, 7))

ax = sns.distplot(y_test, hist=False, color="r", label="Actual Value")
sns.distplot(y_test_pred, hist=False, color="b", label="predicted Values" , ax=ax)
plt.title('Actual vs predicted Values for TZ_AC_MLGL_correl_tx ')
plt.show()
plt.close()
```
